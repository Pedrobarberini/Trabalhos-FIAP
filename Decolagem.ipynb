```python
def verificar_decolagem(telemetria):
    print(f"--- Iniciando Verificação do Cenário: {telemetria['nome_cenario']} ---")
    
    status_decolagem = True
    mensagens_erro = []

    # 1. Leitura e Verificação dos Dados
    if not (18 <= telemetria['temp_interna'] <= 26):
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Interna ({telemetria['temp_interna']}°C) fora do padrão.")

    if not (-10 <= telemetria['temp_externa'] <= 35):
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Externa ({telemetria['temp_externa']}°C) extrema.")

    if telemetria['integridade'] != 1:
        status_decolagem = False
        mensagens_erro.append("Erro: Integridade do casco comprometida (Valor diferente de 1).")

    if telemetria['energia'] < 95:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Nível de energia insuficiente ({telemetria['energia']}%).")

    if not (90 <= telemetria['pressao_tanques'] <= 100):
        status_decolagem = False
        mensagens_erro.append(f"Erro: Pressão dos tanques inadequada ({telemetria['pressao_tanques']}%).")

    if telemetria['modulos_criticos'] != "OK":
        status_decolagem = False
        mensagens_erro.append(f"Erro: Falha em módulos críticos (Status: {telemetria['modulos_criticos']}).")

    # 2. Resultado Final Impresso
    if status_decolagem:
        print(">> RESULTADO FINAL: PRONTO PARA DECOLAR 🚀\n")
    else:
        print(">> RESULTADO FINAL: DECOLAGEM ABORTADA 🛑")
        print("Motivos do aborto:")
        for erro in mensagens_erro:
            print(f" - {erro}")
        print("\n")


# Cenário 1: Condições Perfeitas
cenario_ideal = {
    'nome_cenario': 'Dia de Lançamento Ideal',
    'temp_interna': 28,
    'temp_externa': 28,
    'integridade': 1,
    'energia': 98,
    'pressao_tanques': 95,
    'modulos_criticos': "OK"
}

# Cenário 2: Condições de Risco
cenario_risco = {
    'nome_cenario': 'Tempestade e Vazamento',
    'temp_interna': 30,         # Erro: Muito quente
    'temp_externa': -15,        # Erro: Muito frio
    'integridade': 0,           # Erro: Integridade comprometida
    'energia': 30,              # Erro: Bateria baixa
    'pressao_tanques': 130,      # Erro: Pressão alta
    'modulos_criticos': "FALHA" # Erro: Módulos com problema
}

# Executando as verificações
verificar_decolagem(cenario_ideal)
verificar_decolagem(cenario_risco)
```
