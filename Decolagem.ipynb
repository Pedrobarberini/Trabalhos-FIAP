```python
def verificar_decolagem(telemetria):
    print(f"--- Iniciando Verificação do Cenário: {telemetria['nome_cenario']} ---")
    
    status_decolagem = True
    mensagens_erro = []

    # 1.Leitura e Verificação dos Dados
    if telemetria['temp_interna'] < 18:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Interna ({telemetria['temp_interna']}°C) muito baixa. Recomendado entre 18°C e 26°C.")
    elif telemetria['temp_interna'] > 26:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Interna ({telemetria['temp_interna']}°C) muito alta. Recomendado entre 18°C e 26°C.")
    if telemetria['temp_externa'] < -10:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Externa ({telemetria['temp_externa']}°C) muito baixa. Recomendado entre -10°C e 35°C.")
    elif telemetria['temp_externa'] > 35:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Temperatura Externa ({telemetria['temp_externa']}°C) muito alta. Recomendado entre -10°C e 35°C.")
    if telemetria['integridade'] != 1:
        status_decolagem = False
        mensagens_erro.append("Erro: Integridade do casco comprometida (Valor diferente de 1).")
    if telemetria['energia'] < 95:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Nível de energia insuficiente ({telemetria['energia']}%). Recomendado acima de 95%.")
    if telemetria['pressao_tanques'] < 90:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Pressão dos tanques muito baixa ({telemetria['pressao_tanques']}%). Recomendado entre 90% e 100%.")
    elif telemetria['pressao_tanques'] > 100:
        status_decolagem = False
        mensagens_erro.append(f"Erro: Pressão dos tanques muito alta ({telemetria['pressao_tanques']}%). Recomendado entre 90% e 100%.")
    if telemetria['modulos_criticos'] != "OK":
        status_decolagem = False
        mensagens_erro.append(f"Erro: Falha em módulos críticos (Status: {telemetria['modulos_criticos']}) Modulo Crítico.")

    # 2.Resultado Final Impresso
    if status_decolagem:
        print(" RESULTADO FINAL: PRONTO PARA DECOLAR! \n")
    else:
        print(" RESULTADO FINAL: DECOLAGEM ABORTADA!")
        print("Motivos do aborto:")
        for erro in mensagens_erro:
            print(f" - {erro}")
        print("\n")


# Cenário 1: Condições Perfeitas
cenario_ideal = {
    'nome_cenario': 'Dia de Lançamento Ideal',
    'temp_interna': 22,
    'temp_externa': 28,
    'integridade': 1,
    'energia': 98,
    'pressao_tanques': 95,
    'modulos_criticos': "OK"
}

# Cenário 2: Condições de Risco
cenario_risco = {
    'nome_cenario': 'Tempestade e Vazamento',
    'temp_interna': 30,      
    'temp_externa': -15,        
    'integridade': 0,           
    'energia': 30,              
    'pressao_tanques': 130,      
    'modulos_criticos': "FALHA" 
}

# Executando as verificações
verificar_decolagem(cenario_ideal)
verificar_decolagem(cenario_risco)
```
